```python
import pandas as pd
import numpy as np

df = pd.read_csv('./az.csv')
```

```python
prepend_url = 'http://imageweb/images/CPALinks'
df
```

```python
import os

for i, el in enumerate(df['Image_FileName_DAPI']):
    url = prepend_url + df['Image_PathName_DAPI'][i] + '/' + el
    
    compound = str(df['Image_Metadata_Compound'][i])
    concentration = str(df['Image_Metadata_Concentration'][i])
    plate = str(df['Image_Metadata_Plate_DAPI'][i])
    
    if not os.path.exists('./images/' + compound):
        os.makedirs('./images/' + compound)
    
    if not os.path.exists('./images/' + compound + '/' + concentration):
        os.makedirs('./images/' + compound + '/' + concentration)
    
    if not os.path.exists('./images/' + compound + '/' + concentration + '/' + plate):
        os.makedirs('./images/' + compound + '/' + concentration + '/' + plate)
        
    import urllib

    DAPI_image = urllib.URLopener()
    DAPI_image.retrieve(url, './images/' + compound + '/' + concentration + '/' + plate + '/' + el)
    
    

    
```
